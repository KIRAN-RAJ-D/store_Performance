```python
sales_df['Date'] = pd.to_datetime(sales_df['Date'], errors='coerce')
sales_df['Year'] = sales_df['Date'].dt.year
```

```python
# Group by 'Year' and 'Store', and calculate total and average sales
performance_df = sales_df.groupby(['Year', 'Store']).agg(
    total_sales=('Weekly_Sales', 'sum'),
    average_sales=('Weekly_Sales', 'mean')
).reset_index()

# Display the result
performance_df
```

# identify the profermance of the store on week wise with without offers 

```python
sales_df['Year'] = sales_df['Date'].dt.year
sales_df['Week'] = sales_df['Date'].dt.isocalendar().week
```

```python
weekly_performance_df = sales_df.groupby(['Store', 'Year', 'Week', 'IsHoliday']).agg(
    total_sales=('Weekly_Sales', 'sum'),
    average_sales=('Weekly_Sales', 'mean')
).reset_index()
```

```python
holiday_performance = weekly_performance_df[weekly_performance_df['IsHoliday'] == True]
non_holiday_performance = weekly_performance_df[weekly_performance_df['IsHoliday'] == False]


print(weekly_performance_df)
```
